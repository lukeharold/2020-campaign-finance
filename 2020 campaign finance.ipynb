```python
buttigieg.contributor_city.value_counts().reset_index()
```

```python
buttigieg_west_hollywood = buttigieg[buttigieg.contributor_city == "WEST HOLLYWOOD"]
```

```python
buttigieg_california = buttigieg[buttigieg.contributor_state == "CA"]
```

```python
buttigieg_beverly_hills = buttigieg[buttigieg.contributor_city == "BEVERLY HILLS"]
```

```python
buttigieg_los_angeles = buttigieg[buttigieg.contributor_city == "LOS ANGELES"]
```

```python
buttigieg_west_hollywood.contribution_receipt_amount.reset_index().sort_values("contribution_receipt_amount", ascending=False)
```

```python
buttigieg_west_hollywood.groupby(["contributor_first_name", "contributor_last_name"]).contribution_receipt_amount.sum().reset_index().sort_values("contribution_receipt_amount", ascending=False)
```
